<p>The 2020 election was hotly contested and raised many controversies. One of these was the idea that the election was fraudulent and therefore stolen. This small project attempts to do the hard task of "proving a negative" -that it was not fraudulent. The goal here is to show the results from the 2016 and 2020 elections where President Trump won and then lost.</p>
<p>The results and the voter turnout shown here will help to show that the turnout was reasonable. Here an <b>assumption</b> will be made that if a large spike in turnout shows up then maybe manufactured votes were turned in. However, if the turnout (while recognized as universally higher in 2020) was <i>relatively</i> even across the states or comparable to 2016 then there were not manufactured votes in the contested states to create a win for Biden.  **Also, this subtly pokes holes in another issue around the amount(quantity) of votes Trump received meant Biden couldn’t statistically win. These turnouts show otherwise.</p>
<p><b>NOTE:</b> The author does understand that in researching this topic there were reported and documented irregularities or errors in collecting all the votes of the election. (To be expected with the scope of an election of over 150 million people.) However, as even agreed by Matt Braynard -a republican data analyst who found and called voters who had been miscategorized and was also on Trump’s campaign staff- these errors were not enough to overturn the election. That said, it is agreed that these errors and other confusion points around the voting logistical process need to be ironed out to reduce any contention about who is voted into office for any political party.</p>

What is the voting-age population (VAP) and the voting-eligible population (VEP)?<p>
The voting-age population, known by the acronym VAP, is defined by the Bureau of the Census as everyone residing in the United States, age 18 and older. Before 1971, the voting-age population was age 21 and older for most states.</p>
<p></p>
The voting-eligible population or VEP is a phrase the curator of the turnout data coined to describe the population that is eligible to vote. Counted among the voting-age population are persons who are ineligible to vote, such as non-citizens, felons (depending on state law), and mentally incapacitated persons. Not counted are persons in the military or civilians living overseas. *From source:<a>http://www.electproject.org/home/voter-turnout/faq/denominator</a>
<p>For this project we will be using the VEP as it seems more reasonable by removing ineligible voters, and in looking at the numbers the percentages aren't wildly different.</p>

The following two graphs are <b>both interactive</b> and show the voting turnout percentages by state and then color coded by the winning party. Red for Republican and Blue for Democrat. For both graphs the starting point is at 20% to show more variation. The lines are where the average turn outs are for the states won by a particular party. 
<p>For instance here we see in both cases states where Republicans won had lower turn out percentages for both elections.</p>
<p>Most importantly here though there weren't many outstanding spikes in turn out. <b>Especially</b> in states that were contested in the news (MI, WI, NV, GA). This means holding our assumption stated at the beginning, no spikes mean no created or manufactured votes large enough to show up to make a significant case to be fraudulent. This also despite that overall as seen in the averages there was more turn out in 2020 than there was in 2016</p>

```python
tag = HTML('''<script>
code_show=true; 
function code_toggle() {
    if (code_show){
        $('div.cell.code_cell.rendered.selected div.input').hide();
    } else {
        $('div.cell.code_cell.rendered.selected div.input').show();
    }
    code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
This shows code for following graph, click to see it <a href="javascript:code_toggle()">here</a>.''')
display(tag)
#############  Now Code --

statesLs=turnOut2020['State_Abv'].tolist()
tOutLs=turnOut2020['VEP Highest Office'].tolist()
tOutLs= [float(x.strip('%')) for x in tOutLs]

cmap={'REPUBLICAN':'#B22222',
      'DEMOCRAT':'#0000CD'}


p = figure(y_range=statesLs, plot_height=820, title="Voting Turn Out (%) Percentage and Results 2020 by winning color")

repAvg=fullDataVotes[(fullDataVotes['year']==2020) & (fullDataVotes['party_detailed']=='REPUBLICAN')]['VEP Highest Office'].mean()
demAvg=fullDataVotes[(fullDataVotes['year']==2020) & (fullDataVotes['party_detailed']=='DEMOCRAT')]['VEP Highest Office'].mean()

p.hbar(y='State_Abv', right='VEP Highest Office', width=0.5, source=fullDataVotes[fullDataVotes['year']==2020],
       color=factor_cmap('party_detailed', palette=list(cmap.values()), factors=list(cmap.keys())), alpha=.9)

# Set some properties to make the plot look better
p.ygrid.grid_line_color = None
p.x_range.start = 20

p.line(x=[repAvg,repAvg],y=[0,51], line_color="#F08080", line_width=4)
p.line(x=[demAvg,demAvg],y=[0,51], color="#87CEFA", line_width=4)
show(p)
```
